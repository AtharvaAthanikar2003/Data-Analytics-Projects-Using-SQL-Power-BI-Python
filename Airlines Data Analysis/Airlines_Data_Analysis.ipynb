Q2. Show Bar Graphs representing the Departure and Arrival Timings

```python
data.head()
```

```python
# Show no.of Departure time for the flights
data['departure_time'].value_counts()
```

```python
# Show no.of Arrival time for the flights
data['arrival_time'].value_counts()
```

```python
# Show the Departure and Arrival Timings for the flights with their counts
plt.figure(figsize = (16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index, data['departure_time'].value_counts().values, color = ['r','b'])
plt.title("Departure Time")
plt.xlabel("Dept Time")
plt.ylabel("Dept Freq")
plt.subplot(1,2,2)
plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color = ['g','y'])
plt.title("Arrival Time")
plt.xlabel("Arr Time")
plt.ylabel("Arr Freq")
plt.show()
```
